```python

pymoli2=pymoli_df[['SN', 'Gender','Age']]


demo = pymoli2.drop_duplicates()



gender_totals = demo['Gender'].value_counts()


gender_percent = gender_totals/num_p

gender_demo = pd.DataFrame({"total count":gender_totals,"Percentage of Players":gender_percent})


gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map("{:,.2%}".format)

gender_demo

```

```python
# Gender Purchases

gender_purchase_total = pymoli_df.groupby(['Gender']).sum()["Price"].rename("Total Purchase Value")

gender_avg = pymoli_df.groupby(['Gender']).mean()["Price"].rename("Average Purchase Price")

gender_counts = pymoli_df.groupby(['Gender']).count()["Price"].rename("Purchase Count")


norm_total = gender_purchase_total/gender_demo["total count"]

gender_data = pd.DataFrame({"Purchase Count":gender_counts,"Average Purchase Price":gender_avg,"Total Purchase Value":gender_purchase_total,"Normalized Totals":norm_total})

gender_data['Normalized Totals'] = gender_data['Normalized Totals'].map("${:,.2f}".format)

gender_data['Average Purchase Price'] = gender_data['Average Purchase Price'].map("${:,.2f}".format)

gender_data['Total Purchase Value'] = gender_data['Total Purchase Value'].map("${:,.2f}".format)


gender_data
```

```python
#Top 5 spenders
top_spenders = pymoli_df.groupby("SN")


purchases = top_spenders["Purchase ID"].count()

avg_spender = top_spenders["Price"].mean()


total_spent = top_spenders["Price"].sum()

# Place analysis data into a summary DataFrame
top_df = pd.DataFrame({"Purchase Count": purchases,
                             "Average Purchase Price": avg_spender,
                             "Total Purchase Value":total_spent})

# Sort values in descending order
top_df = top_df.sort_values(["Purchase Count"], ascending=False).head()

# Convert values to float and format to 2 decimal places 
top_df.style.format({"Purchase Count":"{:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```
